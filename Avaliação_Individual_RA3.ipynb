# Questão 1 - Vale até 1 ponto
Em uma região desértica, pesquisadores estão estudando o crescimento de uma espécie rara de cacto. O crescimento do cacto ao longo do tempo é modelado pela função $y = x^5(2 - x)$, onde $y$ representa a altura do cacto em metros e $x$ representa o tempo em anos desde o início do estudo. Os pesquisadores querem determinar a quantidade total de biomassa acumulada pelo cacto ao longo dos primeiros 2 anos. Sabemos, de estudos anteriores que a biomassa acumulada é proporcional a área sob a curva de crescimento do cacto ao longo do tempo. Determine a área sob a curva da função $y = x^5(2 - x)$ nestes dois anos.



# R1:

Para calcular a área sob a curva, é necessario utilizar a integral definida da função $ y $ no intervalo de 0 a 2:

$$ A = \int_{0}^{2} x^5 (2 - x) \, dx $$

### Passo 1: Expandir a Função

Primeiro, é necessario expandir a função $ x^5 (2 - x) $:

$$ x^5 (2 - x) = 2x^5 - x^6 $$

### Passo 2: Integrar Termo a Termo

integrando termo a termo:

$$ \int_{0}^{2} (2x^5 - x^6) \, dx = \int_{0}^{2} 2x^5 \, dx - \int_{0}^{2} x^6 \, dx $$

Para o primeiro termo:

$$ \int_{0}^{2} 2x^5 \, dx = 2 \int_{0}^{2} x^5 \, dx = 2 \left[ \frac{x^6}{6} \right]_{0}^{2} = 2 \left( \frac{2^6}{6} - \frac{0^6}{6} \right) = 2 \left( \frac{64}{6} \right) = \frac{128}{6} = \frac{64}{3} $$

Para o segundo termo:

$$ \int_{0}^{2} x^6 \, dx = \left[ \frac{x^7}{7} \right]_{0}^{2} = \frac{2^7}{7} - \frac{0^7}{7} = \frac{128}{7} $$

### Passo 3: Calcular a Área Total

Subtraindo os resultados das integrações:

$$ A = \frac{64}{3} - \frac{128}{7} $$

Para realizar a subtração, é necessario colocar os termos com um denominador comum:

$$ A = \frac{64}{3} - \frac{128}{7} = \frac{64 \cdot 7}{3 \cdot 7} - \frac{128 \cdot 3}{7 \cdot 3} = \frac{448}{21} - \frac{384}{21} = \frac{64}{21} $$

Portanto, a área sob a curva da função $ y = x^5 (2 - x) $ nos primeiros 2 anos é:

$$ A = \frac{64}{21} \text{ metros}^2 $$

Código em Python para calcular e visualizar a área sob a curva:



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definindo a variável e a função
x = sp.symbols('x')
y = x**5 * (2 - x)

# Calculando a integral definida
area = sp.integrate(y, (x, 0, 2))
area = area.evalf()

# Exibindo o resultado da integral
print(f"A área sob a curva é: {area} metros²")

# Plotando a função e a área sob a curva
x_vals = np.linspace(0, 2, 400)
y_vals = [float(y.subs(x, val)) for val in x_vals]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$y = x^5 (2 - x)$')
plt.fill_between(x_vals, y_vals, alpha=0.3)
plt.title('Área sob a Curva de Crescimento do Cacto')
plt.xlabel('Tempo (anos)')
plt.ylabel('Altura (metros)')
plt.legend()
plt.grid(True)
plt.show()

```

# Questão 3 - Vale até 1 ponto

A aceleração de um objeto é dada por $a(t) = \cos(\pi t)$, e sua velocidade no tempo $t = 0$ é $\frac{1}{2\pi}$. Usando apenas álgebra, encontre a distância total percorrida por este objeto nos primeiros $1,5$ segundos deste movimento. Dica: antes de calcular qualquer coisa verifique cuidadosamente a equação da aceleração. Não é necessário, mas talvez seja interessante, traçar um gráfico da aceleração e outro da velocidade para entender o problema. É necessário entender o movimento do objeto para saber quanto espaço ele percorreu.

# R3:

### Encontrando $v(t)$

Como $v(0) = \frac{1}{2\pi}$, então:

$$ v(t) = \int_{0}^{t} a(\tau) \, d\tau = \int_{0}^{t} \cos(\pi \tau) \, d\tau $$

$$ v(t) = \frac{1}{\pi} \sin(\pi t) \Bigg|_{0}^{t} = \frac{1}{\pi} \sin(\pi t) - \frac{1}{\pi} \sin(0) = \frac{1}{\pi} \sin(\pi t) $$

### Encontrando $x(t)$

Como $x(0) = 0$, então:

$$ x(t) = \int_{0}^{t} v(\tau) \, d\tau = \int_{0}^{t} \frac{1}{\pi} \sin(\pi \tau) \, d\tau $$

$$ x(t) = -\frac{1}{\pi^2} \cos(\pi t) \Bigg|_{0}^{t} = -\frac{1}{\pi^2} \cos(\pi t) + \frac{1}{\pi^2} \cos(0) = -\frac{1}{\pi^2} \cos(\pi t) + \frac{1}{\pi^2} $$

### Calculando a Distância Total Percorrida

Para encontrar a distância total percorrida nos primeiros $1,5$ segundos, é necessario avaliar $x(t)$ em $t = 1,5$:

$$ x(1.5) = -\frac{1}{\pi^2} \cos(\frac{3\pi}{2}) + \frac{1}{\pi^2} $$

$$ x(1.5) = -\frac{1}{\pi^2} \cdot 0 + \frac{1}{\pi^2} = \frac{1}{\pi^2} $$

Portanto, a distância total percorrida nos primeiros $1,5$ segundos deste movimento é $\frac{1}{\pi^2}$ metros.

## Análise dos Gráficos

### Gráfico da Aceleração em Função do Tempo

O gráfico da aceleração em função do tempo é uma curva oscilatória representada pela função $ a(t) = \cos(\pi t) $. Inicialmente, a aceleração é máxima (positiva) quando $ t = 0 $, atingindo o valor de $ 1 $ $ m/s^2 $, e diminui continuamente até se tornar máxima (negativa) em $ t = 0.5 $ segundos, atingindo o valor de $ -1 $ $ m/s^2 $. Após esse ponto, a aceleração volta a aumentar e atinge $ 0 $ $ m/s^2 $ em $ t = 1 $ segundo. A partir desse ponto, a aceleração continua oscilando entre $ -1 $ e $ 1 $ $ m/s^2 $ com um período de $ 1 $ segundo.

### Gráfico da Velocidade em Função do Tempo

O gráfico da velocidade em função do tempo é uma curva senoidal representada pela função $ v(t) = \frac{1}{\pi} \sin(\pi t) $. Inicialmente, a velocidade é positiva e aumenta rapidamente, atingindo o valor máximo de $ \frac{1}{\pi} $ $ m/s $ em $ t = 0.25 $ segundos. Em seguida, a velocidade diminui até se tornar $ 0 $ $ m/s $ em $ t = 0.5 $ segundos. Após esse ponto, a velocidade se torna negativa e atinge o valor mínimo de $ -\frac{1}{\pi} $ $ m/s $ em $ t = 0.75 $ segundos. A partir desse ponto, a velocidade continua oscilando entre $ -\frac{1}{\pi} $ e $ \frac{1}{\pi} $ $ m/s $ com um período de $ 1 $ segundo.







```python
import numpy as np
import matplotlib.pyplot as plt

# Função para a aceleração
def acceleration(t):
    return np.cos(np.pi * t)

# Função para a velocidade
def velocity(t):
    return (1 / np.pi) * np.sin(np.pi * t)

# Intervalo de tempo de 0 a 1.5 segundos
t_vals = np.linspace(0, 1.5, 400)

# Calculando os valores da aceleração e velocidade
a_vals = acceleration(t_vals)
v_vals = velocity(t_vals)

# Plotando o gráfico da aceleração
plt.figure(figsize=(10, 6))
plt.plot(t_vals, a_vals, label='Aceleração ($a(t) = \cos(\pi t)$)', color='blue')
plt.title('Gráfico da Aceleração em Função do Tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('Aceleração ($m/s^2$)')
plt.legend()
plt.grid(True)
plt.show()

# Plotando o gráfico da velocidade
plt.figure(figsize=(10, 6))
plt.plot(t_vals, v_vals, label='Velocidade ($v(t) = \frac{1}{\pi} \sin(\pi t)$)', color='red')
plt.title('Gráfico da Velocidade em Função do Tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (m/s)')
plt.legend()
plt.grid(True)
plt.show()
```

Use o método dos discos para determinar o volume do sólido obtido pela rotação da região delimitada por $y = \sqrt{x}$, $y = 3$ e o próprio eixo $y$ em torno do eixo $y$.


# R5:

Para determinar o volume do sólido utilizando o método de Lagrange, é necessario considerar a região delimitada pelas curvas $y = \sqrt{x}$ e $y = 3$ no plano $xy$. O volume $V$ do sólido é dado pela integral definida:

$$ V = \pi \int_{0}^{3} [f(y)]^2 \, dy $$

onde $f(y)$ é a função que descreve a distância da curva ao eixo de rotação. Neste caso, $f(y) = \sqrt{y}$.

Substituindo na fórmula:

$$ V = \pi \int_{0}^{3} (\sqrt{y})^2 \, dy $$

Calculando a integral:

1. **Integral Indefinida**:
   $$ \int y \, dy = \frac{y^2}{2} + C $$
   
2. **Substituição dos Limites de Integração**:
   $$ \left[ \frac{y^2}{2} \right]_{0}^{3} $$
   Substituindo $ y = 3 $:
   $$ \frac{3^2}{2} = \frac{9}{2} $$
   Substituindo $ y = 0 $:
   $$ \frac{0^2}{2} = 0 $$
   
3. **Diferença entre os Limites**:
   $$ \frac{9}{2} - 0 = \frac{9}{2} $$
   
4. **Multiplicação por $ \pi $**:
   $$ V = \pi \left( \frac{9}{2} \right) = \frac{9\pi}{2} $$

Portanto, o volume do sólido de revolução é $ \frac{9\pi}{2} $ unidades cúbicas.

No gráfico da função $f(x)$, é possivel observar que é uma função decrescente no intervalo $[0, 2]$. A função é composta por duas partes: uma parte linear ($8x - 3$) e uma parte exponencial ($5e^{2-x}$). A inclinação da linha é positiva devido ao termo linear, enquanto a parte exponencial contribui para o decaimento da função. A média da função no intervalo $[0, 2]$ pode ser visualizada como a linha horizontal que passa pelo ponto médio do intervalo.

# (b):
No gráfico da função $x(t)$, é possivel observar a curva que representa a posição do objeto ao longo do tempo. A função é uma equação cúbica que descreve o movimento do objeto. A velocidade média do objeto no intervalo de tempo de $t = 1$ segundo até $t = 4$ segundos pode ser visualizada como a inclinação da linha secante que une os pontos $(1, x(1))$ e $(4, x(4))$.




```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função
def f(x):
    return 8*x - 3 + 5*np.exp(2 - x)

# Gerando valores de x
x_values = np.linspace(0, 2, 400)

# Calculando os valores de y usando a função
y_values = f(x_values)

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = 8x - 3 + 5e^{2-x}$', color='blue')
plt.title('Gráfico da Função $f(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função de deslocamento
def x(t):
    return t**3 - 6*t**2 + 9*t + 1

# Gerando valores de t
t_values = np.linspace(0, 5, 400)

# Calculando os valores de x usando a função de deslocamento
x_values = x(t_values)

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(t_values, x_values, label='$x(t) = t^3 - 6t^2 + 9t + 1$', color='red')
plt.title('Gráfico da Função $x(t)$')
plt.xlabel('t (segundos)')
plt.ylabel('x (metros)')
plt.grid(True)
plt.legend()
plt.show()
```

Nos ítens a, b a seguir, encontre a família de antiderivadas das seguintes funções:

a. $f(x) = 4x^3 - 6x^2 + 2x - 5 $.

b. $f(x) = \frac{1}{x} + e^x$ para todo $x\neq 0$.




# R7:

# A):

Para encontrar a família de antiderivadas da função $f(x) = 4x^3 - 6x^2 + 2x - 5$, é necessario calcular a integral indefinida de $f(x)$:

$$
\int f(x) \, dx = \int (4x^3 - 6x^2 + 2x - 5) \, dx = x^4 - 2x^3 + x^2 - 5x + C,
$$

onde $C$ é uma constante de integração. Portanto, a família de antiderivadas de $f(x)$ é $F(x) = x^4 - 2x^3 + x^2 - 5x + C$, onde $C$ é uma constante arbitrária.

# B):

Para encontrar a família de antiderivadas da função $f(x) = \frac{1}{x} + e^x$ para $x \neq 0$, é necessario calcular a integral indefinida de $f(x)$:

$$
\int f(x) \, dx = \int \left( \frac{1}{x} + e^x \right) \, dx = \ln|x| + e^x + C,
$$

onde $C$ é uma constante de integração. Portanto, a família de antiderivadas de $f(x)$ é $F(x) = \ln|x| + e^x + C$, onde $C$ é uma constante arbitrária.
