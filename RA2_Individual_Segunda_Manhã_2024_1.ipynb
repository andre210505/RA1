#QUESTÃO 3:

Encontrando os pontos críticos de $ f(x) = e^{2x} - e^x$:


$$f'(x) = \frac{d}{dx} (e^{2x} - e^x) = 2e^{2x} - e^x$$

Agora, igualamos a derivada a zero para encontrar os pontos críticos:


$$2e^{2x} - e^x = 0$$


$$e^x(2e^x - 1) = 0$$

Isso implica que $e^x = 0$ ou $2e^x - 1 = 0 $. No entanto, $e^x$ nunca é zero, então a única solução é:


$$2e^x - 1 = 0 \implies e^x = \frac{1}{2} \implies x = -\ln(2)$$

Para determinar a natureza deste ponto crítico, encontramos a segunda derivada $ f''(x) $:

$$f''(x) = \frac{d}{dx} (2e^{2x} - e^x) = 4e^{2x} - e^x$$

Avaliando a segunda derivada no ponto crítico $ x = -\ln(2)$:


$$f''(-\ln(2)) = 4e^{2(-\ln(2))} - e^{-\ln(2)}$$

$$f''(-\ln(2)) = 4e^{-2\ln(2)} - \frac{1}{2}$$

$$f''(-\ln(2)) = 4(2^{-2}) - \frac{1}{2}$$

$$ f''(-\ln(2)) = \frac{4}{4} - \frac{1}{2}$$

$$f''(-\ln(2)) = 1 - \frac{1}{2}$$

$$f''(-\ln(2)) = \frac{1}{2}$$

Como $ f''(-\ln(2)) > 0 $, concluímos que $ x = -\ln(2) $ é um ponto de mínimo local.

Para a função $ f(x) = e^x + x $, a primeira derivada é $ f'(x) = e^x + 1 $. Igualando a zero:

$$e^x + 1 = 0 $$

Isso não possui solução real, portanto não existem pontos críticos reais para $f(x) = e^x + x $.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from ipywidgets import interact, fixed
from IPython.display import display

# Definindo a função a ser analisada
def f(x):
    return np.exp(2*x) - np.exp(x)

# Encontrando o ponto crítico da função
x = sp.Symbol('x')
df = sp.diff(sp.exp(2*x) - sp.exp(x), x)
x_critico_value = sp.solve(df, x)

# Convertendo os valores simbólicos para valores numéricos
x_critico_value = [float(val) for val in x_critico_value]

# Criando um gráfico interativo
def plot_function(x_critico):
    x_vals = np.linspace(-2, 2, 400)
    y_vals = f(x_vals)

    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label='$f(x) = e^{2x} - e^x$')
    plt.scatter(x_critico, f(x_critico), color='red', label='Ponto Crítico')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Gráfico da função $f(x) = e^{2x} - e^x$ com ponto crítico')
    plt.legend()
    plt.grid(True)
    plt.show()

# Exibindo o gráfico interativo
interact(plot_function, x_critico=x_critico_value)

```

Análise:

Comportamento da Função: A partir do gráfico, observamos que a função é decrescente no intervalo considerado, ou seja, à medida que $x$ aumenta, $f(x)$ diminui.

Ponto Crítico: O ponto crítico da função é representado no gráfico por um ponto vermelho. Este ponto corresponde ao valor onde a derivada da função se anula, indicando uma possível mudança no comportamento da função. No caso da função $f(x) = e^{2x} - e^x$, o ponto crítico está localizado em $x \approx -0.693$.
    
Interpretação do Ponto Crítico: Como a função é decrescente antes do ponto crítico e crescente depois dele, o ponto crítico representa um ponto de mínimo local. Isso significa que a função atinge seu menor valor em $x \approx -0.693$ no intervalo considerado.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

# Função para calcular o tempo total T(x)
def tempo_total(x):
    largura_rio = 10
    velocidade_remar = 2
    velocidade_caminhar = 4

    # Calcula a distância remada usando o teorema de Pitágoras
    distancia_remadax = np.sqrt(x**2 + largura_rio**2)

    # Tempo de remada é a distância remada dividida pela velocidade de remar
    tempo_remar = distancia_remadax / velocidade_remar

    # Tempo de caminhada é a distância que resta dividida pela velocidade de caminhar
    distancia_caminhar = largura_rio - x
    tempo_caminhar = distancia_caminhar / velocidade_caminhar

    # Tempo total é a soma do tempo de remada e tempo de caminhada
    tempo_total = tempo_remar + tempo_caminhar

    return tempo_total

# Encontrar o valor de x que minimiza o tempo total
resultado = minimize_scalar(tempo_total)

# Imprimir o resultado
x_minimo = resultado.x
tempo_minimo = resultado.fun
print(f"Para minimizar o tempo, o pescador deve caminhar aproximadamente {x_minimo:.2f} metros antes de começar a remar.")
print(f"O tempo mínimo necessário para atravessar o rio é aproximadamente {tempo_minimo:.2f} segundos.")

# Plotar o gráfico
x_vals = np.linspace(0, 10, 100)
y_vals = tempo_total(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Tempo Total')
plt.scatter(x_minimo, tempo_minimo, color='red', label='Mínimo')
plt.xlabel('Distância caminhada antes de remar (metros)')
plt.ylabel('Tempo Total (segundos)')
plt.title('Tempo Total para Atravessar o Rio')
plt.legend()
plt.grid(True)
plt.show()



```

$$
\frac{\partial f}{\partial x} = \frac{\partial}{\partial x} (x^2y) + \frac{\partial}{\partial x} (2x) + \frac{\partial}{\partial x} (y^3) \\
= 2xy + 2 + 0 \\
= 2xy + 2
$$

$$
\frac{\partial f}{\partial y} = \frac{\partial}{\partial y} (x^2y) + \frac{\partial}{\partial y} (2x) + \frac{\partial}{\partial y} (y^3) \\

= x^2 + 3y^2
$$

Portanto, as derivadas parciais de $f(x, y)$ são:

$$
